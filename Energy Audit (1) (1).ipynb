```python
X=data.iloc[:,[2,3,11,28]].values 
y=data.iloc[:,[32]].values 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0) 
```

```python
X_train
```

```python
from sklearn.linear_model.logistic import LogisticRegression
lgsr=LogisticRegression(random_state=1)
lgsr.fit(X_train,y_train)
```

```python
y_pred=lgsr.predict(X_test) 
y_pred
```

```python
from sklearn.metrics import confusion_matrix 
conf_matrix=confusion_matrix(y_pred,y_test)
conf_matrix
```

```python
from sklearn.metrics import accuracy_score 
acc_sco=accuracy_score(y_test,y_pred) 
acc_sco
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc_clf=DecisionTreeClassifier() 
dtc_clf.fit(X_train,y_train)
```

```python
dtc_y_test=dtc_clf.predict(X_test) 
dtc_y_test 
```

```python
dtc_cm_test=confusion_matrix(y_test,dtc_y_test)
dtc_cm_test 
```
